```python
path = "Forms/death.pdf"
```

```python
from pdf2image import convert_from_path
import cv2
import numpy as np
```

```python
pages = convert_from_path(path, dpi=300, poppler_path=r"poppler-25.07.0/Library/bin")  # remove poppler_path if on macOS/Linux
page = pages[0]
page
```

```python
open_cv_image = cv2.cvtColor(np.array(page), cv2.COLOR_RGB2BGR)
open_cv_image
```
