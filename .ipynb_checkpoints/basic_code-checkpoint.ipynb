The basic code that klind of does wht we want it to do.

```python
import cv2
import numpy as np
from pdf2image import convert_from_path
import matplotlib.pyplot as plt

class PDFProcessor:
    """Simple class to load PDF and detect features."""
    
    def __init__(self, file_path, dpi=300, poppler_path=r"poppler-25.07.0/Library/bin"):
        """
        Initialize with PDF file path.
        Args:
            file_path: Path to the PDF file.
            dpi: DPI for conversion.
        """
        self.file_path = file_path
        self.dpi = dpi
        self.poppler_path = poppler_path
        print(f"Loading PDF from: {file_path}")
        pil_image = convert_from_path(file_path, dpi=dpi, poppler_path=r"poppler-25.07.0/Library/bin")[0]  # First page only
        self.image_np = np.array(pil_image)
        self.gray = cv2.cvtColor(self.image_np, cv2.COLOR_BGR2GRAY) if len(self.image_np.shape) == 3 else self.image_np
        _, self.thresh = cv2.threshold(self.gray, 240, 255, cv2.THRESH_BINARY_INV)

    def _remove_duplicates(self, rectangles, threshold=10):
        """Remove duplicate rectangles based on position threshold."""
        if not rectangles:
            return rectangles
        unique = []
        for curr in rectangles:
            x1, y1, w1, h1 = curr
            if not any(abs(x1 - x2) <= threshold and abs(y1 - y2) <= threshold for x2, y2, w2, h2 in unique):
                unique.append(curr)
        return unique

    def detect_checkboxes(self, target_area=1000, tolerance=400, text_pixel_threshold=0.20):
        """Detect unfilled checkboxes."""
        contours, _ = cv2.findContours(self.thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        rects = []
        for cnt in contours:
            area = cv2.contourArea(cnt)
            if (target_area - tolerance) <= area <= (target_area + tolerance):
                x, y, w, h = cv2.boundingRect(cnt)
                roi = self.thresh[y:y + h, x:x + w]
                non_zero_ratio = cv2.countNonZero(roi) / (w * h)
                if non_zero_ratio < text_pixel_threshold:
                    rects.append((x, y, w, h))
        return self._remove_duplicates(rects)

    def detect_lines(self, checkboxes=None, thickness_thresh=3, y_proximity=50):
        """Detect lines, optionally filtered near checkboxes."""
        contours, _ = cv2.findContours(self.thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        lines = []
        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            length = max(w, h)
            thickness = min(w, h)
            if length >= 1 and thickness <= thickness_thresh:
                lines.append((x, y, w, h))
        
        if checkboxes is None:
            return self._remove_duplicates(lines)
        
        filtered = []
        for line in lines:
            _, line_y, _, _ = line
            y_occurrences = sum(1 for _, cb_y, _, _ in checkboxes if abs(line_y - cb_y) <= y_proximity)
            if not (1 <= y_occurrences <= 4):
                filtered.append(line)
        return self._remove_duplicates(filtered)

    def detect_whitespaces(self, target_area=1000):
        """Detect whitespace boxes."""
        contours, _ = cv2.findContours(self.thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        boxes = []
        for cnt in contours:
            area = cv2.contourArea(cnt)
            x, y, w, h = cv2.boundingRect(cnt)
            length = max(w, h)
            thickness = min(w, h)
            if length > 50 and thickness > 40:
                check_height = 30
                check_y_start = max(0, y - check_height)
                check_y_end = y
                region_above = self.thresh[check_y_start:check_y_end, x:x + w]
                white_pixels = cv2.countNonZero(region_above)
                if 50 < white_pixels < 2000:
                    boxes.append((x, y, w, h))
        
        # IQR filtering on heights
        heights = [h for _, _, _, h in boxes]
        if heights:
            q1, q3 = np.percentile(heights, 25), np.percentile(heights, 75)
            iqr = q3 - q1
            lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
            boxes = [box for box in boxes if lower <= box[3] <= upper]
        
        return self._remove_duplicates(boxes)

    def draw_checkboxes(self, checkboxes):
        """Draw detected checkboxes."""
        result = self.image_np.copy()
        for x, y, w, h in checkboxes:
            cv2.rectangle(result, (x, y), (x + w, y + h), (0, 255, 0), 2)
        return cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

    def draw_lines(self, lines, box_height=30, box_padding=5):
        """Draw info boxes above lines."""
        result = self.image_np.copy()
        for x, y, w, h in lines:
            x1 = max(x - box_padding, 0)
            x2 = min(x + w + box_padding, result.shape[1] - 1)
            y1 = max(y - box_height, 0)
            y2 = y
            cv2.rectangle(result, (x1, y1), (x2, y2), (0, 255, 0), 2)
        return cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

    def draw_whitespaces(self, boxes, shrink=10):
        """Draw shrunk boxes toward the bottom."""
        result = self.image_np.copy()
        for x, y, w, h in boxes:
            ignore_top = int(h * 0.50)
            new_x, new_y = x + shrink, y + ignore_top
            new_w, new_h = w - 2 * shrink, h - ignore_top - shrink
            if new_w > 0 and new_h > 0:
                cv2.rectangle(result, (new_x, new_y), (new_x + new_w, new_y + new_h), (0, 255, 0), 2)
        return cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
```

```python
import cv2
import numpy as np
from pdf2image import convert_from_path
from typing import List, Dict, Tuple, Optional
from dataclasses import dataclass
from sklearn.cluster import DBSCAN
import warnings

@dataclass
class FormElement:
    """Represents a detected form element with metadata."""
    x: int
    y: int
    w: int
    h: int
    element_type: str
    confidence: float
    area: float
    aspect_ratio: float
    
    def to_dict(self):
        return {
            'bbox': (self.x, self.y, self.w, self.h),
            'type': self.element_type,
            'confidence': self.confidence,
            'area': self.area,
            'aspect_ratio': self.aspect_ratio
        }

class AdaptivePDFProcessor:
    """Generalizable PDF form processor that adapts to different form layouts."""
    
    def __init__(self, file_path: str, dpi: int = 300, poppler_path=r"poppler-25.07.0/Library/bin"):
        """
        Initialize with PDF file path and auto-calibrate parameters.
        
        Args:
            file_path: Path to the PDF file
            dpi: DPI for conversion (higher = better quality but slower)
            poppler_path: Path to poppler binaries if needed
        """
        self.file_path = file_path
        self.dpi = dpi
        self.poppler_path = poppler_path
        
        print(f"Loading PDF from: {file_path}")
        
        # Convert PDF to image
        convert_kwargs = {'dpi': dpi}
        if poppler_path:
            convert_kwargs['poppler_path'] = poppler_path
        
        pil_image = convert_from_path(file_path, **convert_kwargs)[0]
        self.image_np = np.array(pil_image)
        
        # Store page dimensions for relative measurements
        self.page_height, self.page_width = self.image_np.shape[:2]
        self.page_area = self.page_height * self.page_width
        
        # Convert to grayscale
        if len(self.image_np.shape) == 3:
            self.gray = cv2.cvtColor(self.image_np, cv2.COLOR_RGB2GRAY)
        else:
            self.gray = self.image_np
        
        # Adaptive thresholding
        self.thresh = self._adaptive_threshold()
        
        # Auto-calibrate parameters based on document
        self.params = self._calibrate_parameters()
        print(f"Auto-calibrated parameters: {self.params}")
    
    def _adaptive_threshold(self) -> np.ndarray:
        """Apply adaptive thresholding that works across different forms."""
        # Try Otsu's method first
        _, thresh_otsu = cv2.threshold(self.gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
        
        # Also try adaptive threshold
        thresh_adaptive = cv2.adaptiveThreshold(
            self.gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
            cv2.THRESH_BINARY_INV, 11, 2
        )
        
        # Use combination approach
        thresh_combined = cv2.bitwise_and(thresh_otsu, thresh_adaptive)
        
        # Morphological operations to clean up
        kernel = np.ones((2, 2), np.uint8)
        thresh_clean = cv2.morphologyEx(thresh_combined, cv2.MORPH_CLOSE, kernel)
        
        return thresh_clean
    
    def _calibrate_parameters(self) -> Dict:
        """Auto-calibrate detection parameters based on the document."""
        contours, _ = cv2.findContours(self.thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        
        # Collect statistics about all contours
        areas = []
        aspect_ratios = []
        widths = []
        heights = []
        
        for cnt in contours:
            area = cv2.contourArea(cnt)
            if area > 10:  # Ignore tiny noise
                x, y, w, h = cv2.boundingRect(cnt)
                areas.append(area)
                aspect_ratios.append(w / h if h > 0 else 0)
                widths.append(w)
                heights.append(h)
        
        # Calculate relative measurements
        areas = np.array(areas)
        aspect_ratios = np.array(aspect_ratios)
        widths = np.array(widths)
        heights = np.array(heights)
        
        # Detect typical checkbox size using clustering
        checkbox_area = self._estimate_checkbox_size(areas)
        
        # Detect typical line thickness
        line_thickness = self._estimate_line_thickness(widths, heights, aspect_ratios)
        
        # Detect text height (helps with relative measurements)
        text_height = self._estimate_text_height(heights)
        
        return {
            'checkbox_area_min': checkbox_area * 0.5,
            'checkbox_area_max': checkbox_area * 1.5,
            'checkbox_aspect_min': 0.7,  # Nearly square
            'checkbox_aspect_max': 1.3,
            'line_thickness_max': line_thickness,
            'line_min_length': text_height * 2,  # Lines should be at least 2x text height
            'text_height': text_height,
            'whitespace_min_width': text_height * 2,
            'whitespace_min_height': text_height * 1.5,
            'duplicate_threshold': max(5, int(text_height * 0.3))
        }
    
    def _estimate_checkbox_size(self, areas: np.ndarray) -> float:
        """Estimate typical checkbox size from area distribution."""
        # Checkboxes are typically in the 0.01% - 0.1% range of page area
        min_area = self.page_area * 0.00001
        max_area = self.page_area * 0.001
        
        checkbox_candidates = areas[(areas >= min_area) & (areas <= max_area)]
        
        if len(checkbox_candidates) > 0:
            # Use clustering to find the most common small square size
            return np.median(checkbox_candidates)
        else:
            # Fallback: estimate based on DPI (assuming ~20x20 pixel checkboxes at 300 DPI)
            scale = self.dpi / 300.0
            return (20 * scale) ** 2
    
    def _estimate_line_thickness(self, widths: np.ndarray, heights: np.ndarray, 
                                  aspect_ratios: np.ndarray) -> int:
        """Estimate typical line thickness from thin contours."""
        # Lines have extreme aspect ratios
        line_mask = (aspect_ratios > 5) | (aspect_ratios < 0.2)
        line_widths = widths[line_mask]
        line_heights = heights[line_mask]
        
        # Thickness is the minimum dimension
        thicknesses = np.minimum(line_widths, line_heights)
        
        if len(thicknesses) > 0:
            return int(np.percentile(thicknesses, 75))
        else:
            scale = self.dpi / 300.0
            return max(2, int(3 * scale))
    
    def _estimate_text_height(self, heights: np.ndarray) -> int:
        """Estimate typical text height from small contours."""
        # Text is typically small contours
        text_candidates = heights[(heights > 5) & (heights < self.page_height * 0.05)]
        
        if len(text_candidates) > 10:
            # Use mode (most common height)
            hist, bins = np.histogram(text_candidates, bins=50)
            mode_idx = np.argmax(hist)
            return int(bins[mode_idx])
        else:
            scale = self.dpi / 300.0
            return int(12 * scale)
    
    def _remove_duplicates(self, rectangles: List[Tuple], threshold: Optional[int] = None) -> List[Tuple]:
        """Remove duplicate rectangles using spatial clustering."""
        if not rectangles or len(rectangles) < 2:
            return rectangles
        
        if threshold is None:
            threshold = self.params['duplicate_threshold']
        
        # Use DBSCAN clustering to group nearby rectangles
        centers = np.array([[x + w/2, y + h/2] for x, y, w, h in rectangles])
        
        if len(centers) < 2:
            return rectangles
        
        clustering = DBSCAN(eps=threshold, min_samples=1).fit(centers)
        
        # Keep the largest rectangle from each cluster
        unique = []
        for label in set(clustering.labels_):
            cluster_rects = [rectangles[i] for i in range(len(rectangles)) 
                           if clustering.labels_[i] == label]
            # Keep the one with largest area
            largest = max(cluster_rects, key=lambda r: r[2] * r[3])
            unique.append(largest)
        
        return unique
    
    def _calculate_confidence(self, element_type: str, area: float, aspect_ratio: float,
                            fill_ratio: float = None) -> float:
        """Calculate confidence score for a detected element."""
        confidence = 1.0
        
        if element_type == 'checkbox':
            # Check if area is in expected range
            expected_area = (self.params['checkbox_area_min'] + self.params['checkbox_area_max']) / 2
            area_score = 1.0 - min(abs(area - expected_area) / expected_area, 1.0)
            
            # Check if aspect ratio is nearly square
            aspect_score = 1.0 - min(abs(aspect_ratio - 1.0), 1.0)
            
            # Check fill ratio (should be mostly empty for unfilled checkboxes)
            fill_score = 1.0
            if fill_ratio is not None:
                fill_score = 1.0 - fill_ratio
            
            confidence = (area_score * 0.4 + aspect_score * 0.4 + fill_score * 0.2)
        
        elif element_type == 'line':
            # Lines should have extreme aspect ratios
            if aspect_ratio > 1:
                aspect_score = min(aspect_ratio / 10.0, 1.0)
            else:
                aspect_score = min((1.0 / aspect_ratio) / 10.0, 1.0)
            confidence = aspect_score
        
        elif element_type == 'text_field':
            # Text fields should have reasonable aspect ratios
            aspect_score = 1.0 - min(abs(aspect_ratio - 4.0) / 4.0, 1.0)
            confidence = aspect_score
        
        return max(0.0, min(1.0, confidence))
    
    def detect_checkboxes(self, min_confidence: float = 0.5) -> List[FormElement]:
        """Detect unfilled checkboxes with confidence scoring."""
        contours, _ = cv2.findContours(self.thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        elements = []
        
        for cnt in contours:
            area = cv2.contourArea(cnt)
            
            # Check if area is in expected range
            if not (self.params['checkbox_area_min'] <= area <= self.params['checkbox_area_max']):
                continue
            
            x, y, w, h = cv2.boundingRect(cnt)
            aspect_ratio = w / h if h > 0 else 0
            
            # Check if nearly square
            if not (self.params['checkbox_aspect_min'] <= aspect_ratio <= self.params['checkbox_aspect_max']):
                continue
            
            # Check fill ratio (should be mostly empty)
            roi = self.thresh[y:y + h, x:x + w]
            fill_ratio = cv2.countNonZero(roi) / (w * h) if (w * h) > 0 else 1.0
            
            if fill_ratio > 0.3:  # Too filled to be an empty checkbox
                continue
            
            confidence = self._calculate_confidence('checkbox', area, aspect_ratio, fill_ratio)
            
            if confidence >= min_confidence:
                elements.append(FormElement(x, y, w, h, 'checkbox', confidence, area, aspect_ratio))
        
        # Remove duplicates
        unique_rects = self._remove_duplicates([(e.x, e.y, e.w, e.h) for e in elements])
        return [e for e in elements if (e.x, e.y, e.w, e.h) in unique_rects]
    
    def detect_lines(self, checkboxes: Optional[List[FormElement]] = None, 
                    min_confidence: float = 0.4) -> List[FormElement]:
        """Detect horizontal and vertical lines, optionally filtered near checkboxes."""
        contours, _ = cv2.findContours(self.thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        elements = []
        
        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            length = max(w, h)
            thickness = min(w, h)
            
            # Check if it's a line (long and thin)
            if length < self.params['line_min_length']:
                continue
            
            if thickness > self.params['line_thickness_max']:
                continue
            
            aspect_ratio = w / h if h > 0 else 0
            area = w * h
            
            confidence = self._calculate_confidence('line', area, aspect_ratio)
            
            if confidence >= min_confidence:
                elements.append(FormElement(x, y, w, h, 'line', confidence, area, aspect_ratio))
        
        # Filter lines near checkboxes if provided
        if checkboxes is not None:
            y_proximity = int(self.params['text_height'] * 2)
            filtered = []
            
            for line in elements:
                # Count how many checkboxes are on approximately the same y-coordinate
                y_occurrences = sum(1 for cb in checkboxes 
                                  if abs(line.y - cb.y) <= y_proximity)
                
                # Keep lines that aren't aligned with 1-4 checkboxes
                # (those are likely checkbox option lines)
                if not (1 <= y_occurrences <= 4):
                    filtered.append(line)
            
            elements = filtered
        
        # Remove duplicates
        unique_rects = self._remove_duplicates([(e.x, e.y, e.w, e.h) for e in elements])
        return [e for e in elements if (e.x, e.y, e.w, e.h) in unique_rects]
    
    def detect_text_fields(self, min_confidence: float = 0.4) -> List[FormElement]:
        """Detect text input fields (white boxes with labels above)."""
        contours, _ = cv2.findContours(self.thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        elements = []
        
        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            
            # Must be reasonably sized
            if w < self.params['whitespace_min_width'] or h < self.params['whitespace_min_height']:
                continue
            
            # Check for text above (indication of a labeled field)
            check_height = int(self.params['text_height'] * 2)
            check_y_start = max(0, y - check_height)
            check_y_end = y
            
            if check_y_start >= check_y_end:
                continue
            
            region_above = self.thresh[check_y_start:check_y_end, x:min(x + w, self.page_width - 1)]
            white_pixels = cv2.countNonZero(region_above)
            
            # Should have some text above (but not too much)
            if not (50 < white_pixels < min(region_above.size * 0.5, 5000)):
                continue
            
            aspect_ratio = w / h if h > 0 else 0
            area = w * h
            
            confidence = self._calculate_confidence('text_field', area, aspect_ratio)
            
            if confidence >= min_confidence:
                elements.append(FormElement(x, y, w, h, 'text_field', confidence, area, aspect_ratio))
        
        # Remove outliers using IQR on heights
        if len(elements) > 4:
            heights = [e.h for e in elements]
            q1, q3 = np.percentile(heights, 25), np.percentile(heights, 75)
            iqr = q3 - q1
            lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
            elements = [e for e in elements if lower <= e.h <= upper]
        
        # Remove duplicates
        unique_rects = self._remove_duplicates([(e.x, e.y, e.w, e.h) for e in elements])
        return [e for e in elements if (e.x, e.y, e.w, e.h) in unique_rects]
    
    def detect_all_elements(self, min_confidence: float = 0.4) -> Dict[str, List[FormElement]]:
        """Detect all form elements at once."""
        checkboxes = self.detect_checkboxes(min_confidence)
        lines = self.detect_lines(checkboxes, min_confidence)
        text_fields = self.detect_text_fields(min_confidence)
        
        return {
            'checkboxes': checkboxes,
            'lines': lines,
            'text_fields': text_fields
        }
    
    def visualize(self, elements: Dict[str, List[FormElement]] = None, 
                  show_confidence: bool = False) -> np.ndarray:
        """Visualize all detected elements with color coding."""
        result = self.image_np.copy()
        
        if elements is None:
            elements = self.detect_all_elements()
        
        # Color scheme
        colors = {
            'checkbox': (0, 255, 0),      # Green
            'line': (255, 0, 0),          # Red
            'text_field': (0, 0, 255)     # Blue
        }
        
        for element_type, element_list in elements.items():
            color = colors.get(element_type, (255, 255, 0))
            
            for elem in element_list:
                if element_type == 'checkbox':
                    # Draw checkbox normally
                    cv2.rectangle(result, (elem.x, elem.y), 
                                (elem.x + elem.w, elem.y + elem.h), color, 2)
                
                elif element_type == 'line':
                    # Draw box above line for text input
                    box_height = int(self.params['text_height'] * 2)
                    box_padding = 5
                    x1 = max(elem.x - box_padding, 0)
                    x2 = min(elem.x + elem.w + box_padding, self.page_width - 1)
                    y1 = max(elem.y - box_height, 0)
                    y2 = elem.y
                    cv2.rectangle(result, (x1, y1), (x2, y2), color, 2)
                
                elif element_type == 'text_field':
                    # Draw shrunk box toward bottom
                    shrink = 10
                    ignore_top = int(elem.h * 0.50)
                    new_x = elem.x + shrink
                    new_y = elem.y + ignore_top
                    new_w = elem.w - 2 * shrink
                    new_h = elem.h - ignore_top - shrink
                    
                    if new_w > 0 and new_h > 0:
                        cv2.rectangle(result, (new_x, new_y), 
                                    (new_x + new_w, new_y + new_h), color, 2)
                
                # Optionally show confidence scores
                if show_confidence:
                    text = f"{elem.confidence:.2f}"
                    cv2.putText(result, text, (elem.x, elem.y - 5),
                              cv2.FONT_HERSHEY_SIMPLEX, 0.4, color, 1)
        
        return cv2.cvtColor(result, cv2.COLOR_BGR2RGB) if len(result.shape) == 3 else result
    
    def export_results(self, elements: Dict[str, List[FormElement]] = None) -> Dict:
        """Export detection results as structured data."""
        if elements is None:
            elements = self.detect_all_elements()
        
        return {
            'page_dimensions': {
                'width': self.page_width,
                'height': self.page_height
            },
            'calibrated_parameters': self.params,
            'elements': {
                element_type: [elem.to_dict() for elem in element_list]
                for element_type, element_list in elements.items()
            },
            'summary': {
                'total_checkboxes': len(elements.get('checkboxes', [])),
                'total_lines': len(elements.get('lines', [])),
                'total_text_fields': len(elements.get('text_fields', []))
            }
        }
```

```python
# Example usage
if __name__ == "__main__":
    # Initialize processor with your PDF
    processor = AdaptivePDFProcessor("Forms/death.pdf", dpi=300)
    
    # Detect all elements
    elements = processor.detect_all_elements(min_confidence=0.005)
    
    # Visualize results
    result_image = processor.visualize(elements, show_confidence=True)
    
    # Export structured data
    results = processor.export_results(elements)
    print(f"Found {results['summary']['total_checkboxes']} checkboxes")
    print(f"Found {results['summary']['total_lines']} lines")
    print(f"Found {results['summary']['total_text_fields']} text fields")
    
```
